# **Competition Code**

## 1. Preprocessing Steps

```python
import json
import os
import re
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import FeatureUnion
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

from scipy.sparse import hstack

DATA_PATH = "./"
```

```python
# Read CSV / JSON files
id_df      = pd.read_csv(f"{DATA_PATH}/data_identification.csv")
emotion_df = pd.read_csv(f"{DATA_PATH}/emotion.csv")

with open(f"{DATA_PATH}/final_posts.json", "r", encoding="utf-8") as f:
    raw_posts = json.load(f)

# Convert to DataFrame: id, text
post_rows = []
for item in raw_posts:
    post = item["root"]["_source"]["post"]
    pid = post["post_id"]
    text = post.get("text", "")
    hashtags = post.get("hashtags", [])

    if hashtags:
        full_text = text + " " + " ".join(f"#{h}" for h in hashtags)
    else:
        full_text = text

    post_rows.append({"id": pid, "text": full_text})

posts_df = pd.DataFrame(post_rows)

# Merge split and emotion info
all_data = posts_df.merge(id_df, on="id", how="left")
all_data = all_data.merge(emotion_df, on="id", how="left")  # test rows have NaN emotion

train_df = all_data[all_data["split"] == "train"].copy()
test_df  = all_data[all_data["split"] == "test"].copy()

print("Train shape:", train_df.shape)
print("Test shape :", test_df.shape)

# Text and labels
train_text_full = train_df["text"].fillna("")
train_labels    = train_df["emotion"]

x_train_text, x_valid_text, y_train, y_valid = train_test_split(
    train_text_full,
    train_labels,
    test_size=0.2,
    random_state=42,
    stratify=train_labels,
)

x_test_text = test_df["text"].fillna("")
```

## 2. Feature Engineering Steps

```python
def build_style_features(text_series: pd.Series) -> np.ndarray:
    
    rows = []
    for text in text_series.astype(str):
        char_len = len(text)
        words = text.split()
        word_len = len(words)

        num_exclam = text.count("!")
        num_qmark  = text.count("?")
        num_dots   = text.count("...")

        num_upper_words = sum(1 for w in words if len(w) >= 2 and w.isupper())

        rows.append([
            np.log1p(char_len),
            np.log1p(word_len),
            num_exclam,
            num_qmark,
            num_upper_words,
            num_dots,
        ])
    return np.array(rows, dtype=np.float32)
```

```python
train_text_clean = x_train_text.fillna("")
valid_text_clean = x_valid_text.fillna("")
test_text_clean  = x_test_text.fillna("")

# Word-level TF-IDF
word_vectorizer = TfidfVectorizer(
    max_features=60000,
    ngram_range=(1, 2),
    sublinear_tf=True,
    min_df=2,
)

# Char-level TF-IDF
char_vectorizer = TfidfVectorizer(
    analyzer="char_wb",
    ngram_range=(3, 5),
    min_df=3,
    max_features=30000,
)

# Combine word + char
tfidf_union = FeatureUnion([
    ("word", word_vectorizer),
    ("char", char_vectorizer),
])

x_train_tfidf = tfidf_union.fit_transform(train_text_clean)
x_valid_tfidf = tfidf_union.transform(valid_text_clean)
x_test_tfidf  = tfidf_union.transform(test_text_clean)

print("TF-IDF shapes (train/valid/test):")
print("  train:", x_train_tfidf.shape)
print("  valid:", x_valid_tfidf.shape)
print("  test :", x_test_tfidf.shape)

# Style features (dense)
style_train = build_style_features(train_text_clean)
style_valid = build_style_features(valid_text_clean)
style_test  = build_style_features(test_text_clean)

print("Style feature shape:", style_train.shape)

# Final features = TF-IDF + style
x_train_feat = hstack([x_train_tfidf, style_train])
x_valid_feat = hstack([x_valid_tfidf, style_valid])
x_test_feat  = hstack([x_test_tfidf,  style_test])

print("Final feature shapes:")
print("  train:", x_train_feat.shape)
print("  valid:", x_valid_feat.shape)
print("  test :", x_test_feat.shape)
```

## 3. Model Implementation Steps

```python
# Logistic Regression
logit_clf = LogisticRegression(
    max_iter=1000,
    solver="lbfgs",
    multi_class="auto",
    C=2.0,              # you can try 1 / 4 etc.
    class_weight="balanced",
)

logit_clf.fit(x_train_feat, y_train)

valid_pred_logit  = logit_clf.predict(x_valid_feat)
valid_proba_logit = logit_clf.predict_proba(x_valid_feat)

macro_f1_logit = f1_score(y_valid, valid_pred_logit, average="macro")
print("Logistic (enhanced features) macro F1:", macro_f1_logit)
print("Logistic valid report:")
print(classification_report(y_valid, valid_pred_logit))

# Multinomial Naive Bayes
nb_clf = MultinomialNB(alpha=0.5)
nb_clf.fit(x_train_feat, y_train)

valid_pred_nb  = nb_clf.predict(x_valid_feat)
valid_proba_nb = nb_clf.predict_proba(x_valid_feat)

macro_f1_nb = f1_score(y_valid, valid_pred_nb, average="macro")
print("Naive Bayes (enhanced features) macro F1:", macro_f1_nb)
print("Naive Bayes valid report:")
print(classification_report(y_valid, valid_pred_nb))

# Soft-voting ensemble: 0.7 * logit + 0.3 * NB
class_labels = logit_clf.classes_
w_logit = 0.7
w_nb    = 0.3

valid_proba_ens = (w_logit * valid_proba_logit + w_nb * valid_proba_nb) / (w_logit + w_nb)
valid_pred_ens_idx = np.argmax(valid_proba_ens, axis=1)
valid_pred_ens = class_labels[valid_pred_ens_idx]

macro_f1_ens = f1_score(y_valid, valid_pred_ens, average="macro")
print("Ensemble (0.7 logit + 0.3 NB) macro F1:", macro_f1_ens)
print("Ensemble valid report:")
print(classification_report(y_valid, valid_pred_ens))

# Choose best model for test prediction
best_f1 = max(macro_f1_logit, macro_f1_nb, macro_f1_ens)
if best_f1 == macro_f1_ens:
    print(">>> Using ENSEMBLE for test prediction")
    test_proba_logit = logit_clf.predict_proba(x_test_feat)
    test_proba_nb    = nb_clf.predict_proba(x_test_feat)
    test_proba_ens   = (w_logit * test_proba_logit + w_nb * test_proba_nb) / (w_logit + w_nb)
    test_pred_idx    = np.argmax(test_proba_ens, axis=1)
    test_pred        = class_labels[test_pred_idx]
    model_tag = "ensemble_feat"
elif best_f1 == macro_f1_logit:
    print(">>> Using LOGISTIC for test prediction")
    test_pred = logit_clf.predict(x_test_feat)
    model_tag = "logit_feat"
else:
    print(">>> Using NAIVE BAYES for test prediction")
    test_pred = nb_clf.predict(x_test_feat)
    model_tag = "nb_feat"

# Build submission file
sample_sub = pd.read_csv(f"{DATA_PATH}/samplesubmission.csv")
id_to_label = dict(zip(test_df["id"], test_pred))
sample_sub["emotion"] = sample_sub["id"].map(id_to_label)

output_file = f"submission_{model_tag}.csv"
sample_sub.to_csv(output_file, index=False)
print("Saved submission file:", output_file)
```
